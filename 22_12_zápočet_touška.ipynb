```python
def euler_step(f, x0, y0, h, n):
    """
    Numerick een ODE pomoc Eulerovy metody.
    
    :param f: Funkce f(x, y) pedstavujc derivaci dy/dx
    :param x0: Poten hodnota x
    :param y0: Poten hodnota y
    :param h: Dlka kroku
    :param n: Poet krok
    :return: Pole hodnot y (vektor)
    """
    if n < 1:
        print("Error: Number of steps n must be >= 1.")
        return None

    # Vytvome pole o velikosti n+1 naplnn nulami
    ys = [0.0] * (n + 1)
    ys[0] = y0
    
    x = x0
    y = y0
    
    for i in range(1, n + 1):
        # Samotn Eulerv krok: y_nov = y_star + h * sklon
        y = y + h * f(x, y)
        x = x + h
        ys[i] = y
        
    return ys

# Pklad pouit:
# f = lambda x, y: x + y  (pro rovnici y' = x + y)
# vysledek = euler_step(f, 0, 1, 0.1, 10)



A = 20     #KRITICK BOD
K = 100     #kapacita nosn 
N0 = 60    #Velikost populace N0
r = 0.5



f = lambda N,n: -r*N*(+-(N/A))*(1-(N/K))


#f = lambda x,y: ????


#Co je Y0 .. Y(X0=Y0)

t = euler_step(f,N0,N0,0.1,50)


print(t)


```
